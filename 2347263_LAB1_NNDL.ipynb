# Importing Neccessary Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Defining a class for SingleLayerPerception

```python
class SingleLayerPerceptron:
    def __init__(self, input_size, learning_rate=0.1, epochs=100):
        self.weights = np.random.rand(input_size)
        self.bias = np.random.rand()
        self.learning_rate = learning_rate
        self.epochs = epochs

    def activation(self, x):
        return 1 if x >= 0 else 0

    def predict(self, inputs):
        summation = np.dot(inputs, self.weights) + self.bias
        return self.activation(summation)

    def train(self, X, y):
        for epoch in range(self.epochs):
            for i in range(len(X)):
                prediction = self.predict(X[i])
                error = y[i] - prediction

                self.weights += self.learning_rate * error * X[i]
                self.bias += self.learning_rate * error
```

# AND GATE

```python
# AND gate truth table
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])

y = np.array([0, 0, 0, 1])

# Initialize and train the perceptron
perceptron = SingleLayerPerceptron(input_size=2)
perceptron.train(X, y)

# Test the model
print("\nTesting the AND gate Perceptron:")
for input_set in X:
    prediction = perceptron.predict(input_set)
    print(f"Input: {input_set}, Predicted Output: {prediction}")


plt.figure(figsize=(6, 4))
for i in range(len(X)):
    if y[i] == 0:
        plt.scatter(X[i][0], X[i][1], color='red', marker='o', label='0')
    else:
        plt.scatter(X[i][0], X[i][1], color='blue', marker='x', label='1')

# Find points for plotting the decision boundary line
x_line = np.linspace(-0.5, 1.5, 100)
y_line = (-perceptron.weights[0] * x_line - perceptron.bias) / perceptron.weights[1]
plt.plot(x_line, y_line, color='green', label='Decision Boundary')

plt.title("AND Gate Decision Boundary")
plt.xlabel("Input 1")
plt.ylabel("Input 2")
plt.xlim([-0.5, 1.5])
plt.ylim([-0.5, 1.5])
plt.legend()
plt.show()
```

## Questions:
1. How do the weights and bias values change during training for the AND gate?

  - The weights and bias change incrementally based on the error between the predicted and expected output for each input pair.
For the AND gate, when the inputs are
[1,1], the perceptron updates its weights positively to increase the likelihood of predicting 1. For all other input pairs (which should output 0), the weights are adjusted to reduce the likelihood of predicting 1.
2. Can the perceptron successfully learn the AND logic with a linear decision boundary?

  - Yes, the perceptron successfully learns the AND logic with a linear decision boundary. The plot shows a clear decision boundary that separates the positive (1) from the negative (0) outputs.

# XOR GATE

```python
# XOR gate truth table
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])

y = np.array([0, 1, 1, 0]) # XOR gate outputs

# Initialize and train the perceptron
perceptron = SingleLayerPerceptron(input_size=2)
perceptron.train(X, y)

# Test the model
print("Testing the XOR gate Perceptron:")
for input_set in X:
    prediction = perceptron.predict(input_set)
    print(f"Input: {input_set}, Predicted Output: {prediction}")


plt.figure(figsize=(6, 4))
for i in range(len(X)):
    if y[i] == 0:
        plt.scatter(X[i][0], X[i][1], color='red', marker='o', label='0')
    else:
        plt.scatter(X[i][0], X[i][1], color='blue', marker='x', label='1')

# Find points for plotting the decision boundary line
x_line = np.linspace(-0.5, 1.5, 100)
y_line = (-perceptron.weights[0] * x_line - perceptron.bias) / perceptron.weights[1]
plt.plot(x_line, y_line, color='green', label='Decision Boundary')

plt.title("XOR Gate Decision Boundary")
plt.xlabel("Input 1")
plt.ylabel("Input 2")
plt.xlim([-0.5, 1.5])
plt.ylim([-0.5, 1.5])
plt.legend()
plt.show()
```

## Questions:
- Why does the Single Layer Perceptron struggle to classify the XOR gate?
  - The Single Layer Perceptron (SLP) is a simple neural network model with only one layer of neurons. It can learn linear decision boundaries, but the XOR gate poses a challenge:
    - XOR Operation: The XOR gate outputs 1 if exactly one of its inputs is 1; otherwise, it outputs 0.
    - Problem: The XOR truth table cannot be separated by a single straight line (linearly separable). Look at the graph you provided: the red and blue points representing XOR inputs cannot be divided by a single line.
    - SLP Limitation: SLPs can only create linear decision boundaries. They can’t capture complex patterns like the XOR gate.


- What modifications can be made to the neural network model to handle the XOR gate correctly?

  - Multi-Layer Perceptron (MLP): Introduce additional layers (hidden layers) between input and output layers. These hidden layers allow for non-linear transformations.
  - Activation Functions: Replace the linear activation function (used in SLP) with non-linear ones like the sigmoid, ReLU (Rectified Linear Unit), or tanh. These functions introduce non-linearity, enabling the network to learn complex relationships.
  - Backpropagation: Train the network using backpropagation to adjust weights and biases. This process fine-tunes the model to fit the XOR data.
  - Architecture: A simple MLP with one hidden layer (containing a few neurons) can handle the XOR gate efficiently.
